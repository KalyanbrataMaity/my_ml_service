### Train ML algorithms

```python
# import libraries
import json                                                                 # for saving preprocessing details
import numpy as np                                                          # data manipulation
import pandas as pd                                                         # data manipulation
from sklearn.model_selection import train_test_split                        # used for data split
from sklearn.preprocessing import LabelEncoder                              # for preprocessing
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier   # for training algorithm
import joblib                                                               # saving trained model
```

#### Loading data

```python
# load data
df = pd.read_csv('/Users/kalyanbrata/Work/my_ml_service/research/income_prediction/data/raw/adult.csv')
col_names = [col.strip() for col in df.columns]
df.columns = col_names

df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

x_cols = [c for c in df.columns if c != 'income']

# set input matrix and target column
X = df[x_cols]
y = df['income']

# show first rows of data
df.head()
```

```python
df.shape
```

```python
# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

#### Data Pre-processing

```python
# fill missing values
train_mode = dict(X_train.mode().iloc[0])
X_train = X_train.fillna(train_mode)
print(train_mode)
```

```python
X_train['native-country'].values
```

```python
# convert categoricals
encoders = {}
for column in ['age', 'workclass', 'education', 'marital-status', 
               'occupation', 'relationship', 'race', 'sex', 'native-country']:
    categorical_convert = LabelEncoder()
    X_train[column] = categorical_convert.fit_transform(X_train[column])
    encoders[column] = categorical_convert
```

```python
encoders
```

```python
X_train
```

### Algorithms training

```python
# train the random forest algorithm
rf = RandomForestClassifier(n_estimators=100)
rf = rf.fit(X_train, y_train)
```

```python
# train the extra trees algorithm
et = ExtraTreesClassifier(n_estimators=100)
et = et.fit(X_train, y_train)
```

```python
# save preprocessing objects and rf algorithm
joblib.dump(train_mode, "../trained_models/train_mode.joblib", compress=True)
joblib.dump(encoders, "../trained_models/encoders.joblib", compress=True)
joblib.dump(rf, "../trained_models/random_forest.joblib", compress=True)
joblib.dump(et, "../trained_models/extra_trees.joblib", compress=True)
```
